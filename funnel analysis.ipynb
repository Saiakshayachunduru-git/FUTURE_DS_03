```python

# 1. IMPORT LIBRARIES
import pandas as pd
import zipfile


# 2. OPEN MAIN ZIP

with zipfile.ZipFile("bank+marketing.zip") as main_zip:
    print("Main ZIP contains:", main_zip.namelist())
    
    # Step 1: Open inner zip (bank.zip)
    with main_zip.open("bank.zip") as inner_zip_file:
        with zipfile.ZipFile(inner_zip_file) as inner_zip:
            print("Inner ZIP contains:", inner_zip.namelist())
            
            # Step 2: Load CSV from inner zip
            df = pd.read_csv(inner_zip.open("bank-full.csv"), sep=';')


# 3. CHECK DATA

print("\nData Loaded Successfully!\n")
print(df.head())

# 4. CLEAN DATA

df['y'] = df['y'].map({'yes': 1, 'no': 0})

# 5. FUNNEL ANALYSIS

total = len(df)
converted = df['y'].sum()

print("\n📊 Funnel:")
print("Total:", total)
print("Converted:", converted)
print("Conversion Rate:", (converted/total)*100)
```
